# Liars' Bench — Data Download

Downloads the [Cadenza-Labs/liars-bench](https://huggingface.co/datasets/Cadenza-Labs/liars-bench) dataset from HuggingFace and saves each split to `data/` as Parquet files.

**Note:** This dataset requires agreeing to conditions on HuggingFace. If you hit an authentication error, run `huggingface-cli login` in your terminal first.

```python
from pathlib import Path

from datasets import load_dataset
```

```python
CONFIGS = [
    "alpaca",
    "convincing-game",
    "gender-secret",
    "harm-pressure-choice",
    "harm-pressure-knowledge-report",
    "insider-trading",
    "instructed-deception",
    "soft-trigger",
]

datasets = {}
for config in CONFIGS:
    datasets[config] = load_dataset("Cadenza-Labs/liars-bench", config)
    print(f"{config}: {datasets[config]}")
```

```python
out_dir = Path("../data/liars_bench")
out_dir.mkdir(parents=True, exist_ok=True)

for config, dataset in datasets.items():
    for split, ds in dataset.items():
        out_path = out_dir / f"{config}__{split}.parquet"
        ds.to_parquet(str(out_path))
        print(f"Saved {config}/{split} ({len(ds)} rows) → {out_path.name}")
```
